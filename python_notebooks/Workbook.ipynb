```python
import pandas as pd
import altair as alt
```

```python
url = 'https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv'
df = pd.read_csv(url)

df.head()
```

```python
data_filtered = df.dropna(subset=["Square Footage", "Year Constructed", "Bldg Status"])

status_selection = alt.selection_point(
    fields=['Bldg Status'],
    bind=alt.binding_select(options=data_filtered['Bldg Status'].unique().tolist(), name='Select Building Status')
)

chart1 = alt.Chart(data_filtered).mark_circle().encode(
    x=alt.X('County:O', title='County'),
    y=alt.Y('Square Footage:Q', title='Square Footage'),
    color='Agency Name:N', 
    size='Square Footage:Q',  
    tooltip=['Location Name', 'Square Footage', 'Bldg Status', 'Agency Name']
).properties(
    title='Geographical Distribution of Buildings by Agency',
    width=700,
    height=400
).add_params(
    status_selection
).transform_filter(
    status_selection
)
```

```python
chart2 = alt.Chart(df).mark_bar().encode(
    x=alt.X('County:O', title='County'),
    y=alt.Y('count():Q', title='Number of Buildings'),
    color='Agency Name:N', 
    tooltip=['Year Constructed', 'count()', 'Agency Name']
).properties(
    title='Construction Trends by County and Agency',
    width=700,
    height=400
)
```
